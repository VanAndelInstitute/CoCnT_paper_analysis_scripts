```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal, mannwhitneyu

# ---- define your groups (using your updated names) ----
rb_samples_1 = [
    "Rb_K4me3_H2_CnT",
    "Rb_K4me3_H2_CoCnT_4X",
    "Rb_K4me3_H2_CoCnT_12X",
    "Rb_K4me3_H2_CoCnT_24X",
]

rb_samples_2 = [
    "Rb_K4me3_H23_CnT",
    "Rb_K4me3_H23_CoCnT_4X",
    "Rb_K4me3_H23_CoCnT_12X",
    "Rb_K4me3_H23_CoCnT_24X",
]

rb_samples_3 = [
    "Rb_K4me3_H24_CnT",
    "Rb_K4me3_H24_CoCnT_4X",
    "Rb_K4me3_H24_CoCnT_12X",
    "Rb_K4me3_H24_CoCnT_24X",
]

ms_samples_1 = [
    "Ms_K4me3_H12_CnT",
    "Ms_K4me3_H12_CoCnT_4X",
    "Ms_K4me3_H12_CoCnT_12X",
    "Ms_K4me3_H12_CoCnT_24X",
]

GROUPS = {
    "Rb_K4me3_H2": rb_samples_1,
    "Rb_K4me3_H23": rb_samples_2,
    "Rb_K4me3_H24": rb_samples_3,
    "Ms_K4me3_H12": ms_samples_1,
}

def bh_fdr(pvals):
    """Benjamini–Hochberg FDR correction. Returns adjusted p-values in original order."""
    pvals = np.asarray(pvals, dtype=float)
    n = pvals.size
    order = np.argsort(pvals)
    ranked = pvals[order]
    adj = ranked * n / (np.arange(1, n + 1))
    # enforce monotonicity
    adj = np.minimum.accumulate(adj[::-1])[::-1]
    adj = np.clip(adj, 0, 1)
    out = np.empty_like(adj)
    out[order] = adj
    return out

def within_group_stats(barcode_counts, group_name, conditions, alternative="two-sided"):
    """
    barcode_counts: dict condition -> pandas Series (barcode -> count), from your pipeline
    conditions: list of condition names to compare within this group
    """
    # keep only conditions that exist and have data
    present = [c for c in conditions if c in barcode_counts and len(barcode_counts[c]) > 0]
    if len(present) < 2:
        return None, None

    data = [barcode_counts[c].values for c in present]

    # omnibus test
    kw_stat, kw_p = kruskal(*data)

    # pairwise tests
    rows = []
    pvals = []
    for i in range(len(present)):
        for j in range(i + 1, len(present)):
            a = barcode_counts[present[i]].values
            b = barcode_counts[present[j]].values

            # Mann–Whitney U (nonparametric)
            u_stat, p = mannwhitneyu(a, b, alternative=alternative)

            # effect size (Cliff's delta-like using U)
            # delta = 2U/(n1*n2) - 1  (ranges [-1,1])
            n1, n2 = len(a), len(b)
            delta = (2 * u_stat) / (n1 * n2) - 1

            rows.append({
                "group": group_name,
                "cond_a": present[i],
                "cond_b": present[j],
                "n_a": n1,
                "n_b": n2,
                "median_a": float(np.median(a)),
                "median_b": float(np.median(b)),
                "U": float(u_stat),
                "p_raw": float(p),
                "effect_delta": float(delta),
            })
            pvals.append(p)

    # multiple testing correction within this group
    p_adj = bh_fdr(pvals)
    for r, pa in zip(rows, p_adj):
        r["p_adj_BH"] = float(pa)

    pairwise_df = pd.DataFrame(rows).sort_values("p_adj_BH")
    omnibus = pd.DataFrame([{
        "group": group_name,
        "k": len(present),
        "conditions": ", ".join(present),
        "kruskal_stat": float(kw_stat),
        "kruskal_p": float(kw_p),
    }])

    return omnibus, pairwise_df

# ---- run all groups ----
all_omnibus = []
all_pairwise = []

for gname, conds in GROUPS.items():
    omni, pair = within_group_stats(barcode_counts, gname, conds, alternative="two-sided")
    if omni is not None:
        all_omnibus.append(omni)
        all_pairwise.append(pair)

omnibus_df = pd.concat(all_omnibus, ignore_index=True) if all_omnibus else pd.DataFrame()
pairwise_df = pd.concat(all_pairwise, ignore_index=True) if all_pairwise else pd.DataFrame()

print("\nOmnibus (Kruskal–Wallis) per group:")
print(omnibus_df.to_string(index=False))

print("\nPairwise (Mann–Whitney U) within each group (BH-adjusted):")
print(pairwise_df.head(30).to_string(index=False))

# Optional saves
# omnibus_df.to_csv("reads_per_cell_omnibus_kruskal.csv", index=False)
# pairwise_df.to_csv("reads_per_cell_pairwise_mwu_BH.csv", index=False)
```

```python
# ────────────────────────────────────────────────────────────────
# FRiP per barcode for "conditions" where each condition = 1+ BED files
# ────────────────────────────────────────────────────────────────

import os
import numpy as np
import pandas as pd
import pybedtools
import matplotlib.pyplot as plt

# ────────────────────────────────────────────────────────────────
# 1) Inputs
# ────────────────────────────────────────────────────────────────

CONDITION_BEDS = {
    "Ms_K4me3_H12_CnT": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K27me3(H12)_CnT_041725.bed.gz",
    ],

    "Ms_K4me3_H12_CoCnT_4X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K4me3(H12)_wRb_K27me3(H1)_4XS_4XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K27me3_H1_4XS_4XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H24_4XS_4XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H23_4XS_4XM_060225.bed.gz",
    ],

    "Ms_K4me3_H12_CoCnT_12X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K4me3(H12)_wRb_K27me3(H1)_12XS_12XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K27me3_H1_12XS_12XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H24_12XS_12XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H23_12XS_12XM_060225.bed.gz",
    ],

    "Ms_K4me3_H12_CoCnT_24X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K4me3(H12)_wRb_K27me3(H1)_24XS_24XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K27me3_H1_24XS_24XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H24_24XS_24XM_060225.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K4me3_H12_wRb_K4me3_H23_24XS_24XM_060225.bed.gz",
    ],

    "Rb_K4me3_H23_CnT": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_CnT_060225.bed.gz",
    ],

    "Rb_K4me3_H23_CoCnT_4X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K4me3_H12_4XS_4XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K27me3_H19_4XS_4XM_060225.bed.gz",
    ],

    "Rb_K4me3_H23_CoCnT_12X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K4me3_H12_12XS_12XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K27me3_H19_12XS_12XM_060225.bed.gz",
    ],

    "Rb_K4me3_H23_CoCnT_24X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K4me3_H12_24XS_24XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K4me3_H23_wMs_K27me3_H19_24XS_24XM_060225.bed.gz",
    ],
}

# Peak set (single bed)
PEAK_BED = "/varidata/research/projects/janssens/CoCnT_Data_Analysis/Combined_K4me3_K27me3_beds/K4me3_H23_H12_CnT_CoCnT_Combined.01.stringent.bed"

# Column indices (0-based)
CHR_COL, START_COL, END_COL, BARCODE_COL = 0, 1, 2, 3

# Optional filter for stability: only keep barcodes with >= N total reads across all files in the condition
MIN_TOTAL_READS_PER_BARCODE = 100   # set 0 to disable

# Plot grouping/order
rb_samples = [
    "Rb_K4me3_H23_CnT",
    "Rb_K4me3_H23_CoCnT_4X",
    "Rb_K4me3_H23_CoCnT_12X",
    "Rb_K4me3_H23_CoCnT_24X",
]
ms_samples = [
    "Ms_K4me3_H12_CnT",
    "Ms_K4me3_H12_CoCnT_4X",
    "Ms_K4me3_H12_CoCnT_12X",
    "Ms_K4me3_H12_CoCnT_24X",
]
colour_map = (
    ["#4D4D4D", "#ADD8E6", "#1E90FF", "#104E8B"] +  # Rb
    ["#4D4D4D", "#FFB3FF", "#FF00FF", "#8B008B"]    # Ms
)

OUT_PNG = (
    "/varidata/research/projects/janssens/CoCnT_Data_Analysis/Figure_Panels/"
    "K4me3_H23_H12_FRIP_per_cell_boxplot_ALLREADS_combinedConditions.png"
)

# ────────────────────────────────────────────────────────────────
# 2) Prepare peak BedTool once
# ────────────────────────────────────────────────────────────────
peak_bt = pybedtools.BedTool(PEAK_BED).sort().merge()

# ────────────────────────────────────────────────────────────────
# 3) Helpers
# ────────────────────────────────────────────────────────────────

def load_bed_mincols(path, usecols=(0,1,2,3)):
    return pd.read_csv(path, sep="\t", header=None, compression="infer", usecols=list(usecols))

def count_total_reads_per_barcode(df, barcode_col=3):
    return df[barcode_col].value_counts()

def count_in_peaks_per_barcode(df, peak_bt, chr_col=0, start_col=1, end_col=2, barcode_col=3):
    # Create BedTool from 4-column df (chr, start, end, barcode)
    bt = pybedtools.BedTool.from_dataframe(df[[chr_col, start_col, end_col, barcode_col]])
    in_peak = bt.intersect(peak_bt, wa=True)
    if len(in_peak) == 0:
        return pd.Series(dtype=int)
    in_df = in_peak.to_dataframe(names=[chr_col, start_col, end_col, barcode_col])
    return in_df[barcode_col].value_counts()

def frip_for_condition(bed_paths):
    """
    Returns list of FRiP values (one per barcode), after summing reads across all bed_paths.
    FRiP = (reads in peaks across all files) / (total reads across all files) for each barcode.
    """
    total_counts = None
    peak_counts = None

    for p in bed_paths:
        df = load_bed_mincols(p, usecols=(CHR_COL, START_COL, END_COL, BARCODE_COL))
        if df.empty:
            continue

        tc = count_total_reads_per_barcode(df, barcode_col=BARCODE_COL)
        pc = count_in_peaks_per_barcode(df, peak_bt,
                                        chr_col=CHR_COL, start_col=START_COL, end_col=END_COL, barcode_col=BARCODE_COL)

        total_counts = tc if total_counts is None else total_counts.add(tc, fill_value=0)
        peak_counts  = pc if peak_counts  is None else peak_counts.add(pc, fill_value=0)

    if total_counts is None or len(total_counts) == 0:
        return []

    total_counts = total_counts.astype(int)
    peak_counts = (peak_counts if peak_counts is not None else pd.Series(dtype=float)).astype(float)

    # Optional barcode filter on total reads (across all combined files)
    if MIN_TOTAL_READS_PER_BARCODE and MIN_TOTAL_READS_PER_BARCODE > 0:
        keep = total_counts[total_counts >= MIN_TOTAL_READS_PER_BARCODE].index
        total_counts = total_counts.loc[keep]
        peak_counts = peak_counts.reindex(keep, fill_value=0)

    # FRiP per barcode (aligned)
    frip = peak_counts.reindex(total_counts.index, fill_value=0).divide(total_counts)

    # Clean numeric + bound
    frip = frip.replace([np.inf, -np.inf], np.nan).dropna()
    frip = frip.clip(0, 1)

    return frip.tolist()

# ────────────────────────────────────────────────────────────────
# 4) Compute FRiP per barcode for each condition (combining files)
# ────────────────────────────────────────────────────────────────
frip_per_condition = {}
for cond, paths in CONDITION_BEDS.items():
    if isinstance(paths, str):
        paths = [paths]
    frip_per_condition[cond] = frip_for_condition(paths)

# ────────────────────────────────────────────────────────────────
# 5) Build data & box-plot (gapped layout)
# ────────────────────────────────────────────────────────────────

data_rb = [frip_per_condition[c] for c in rb_samples]
data_ms = [frip_per_condition[c] for c in ms_samples]

all_data = data_rb + data_ms
group_lengths = [len(data_rb), len(data_ms)]

# positions with 1 blank gap
positions, current, gap = [], 0, 1
for g_len in group_lengths:
    positions.extend(range(current, current + g_len))
    current += g_len + gap

fig, ax = plt.subplots(figsize=(6, 6))
bp = ax.boxplot(
    all_data,
    positions=positions,
    widths=0.8,
    patch_artist=True,
    showfliers=False
)

# color + style
for patch, col in zip(bp["boxes"], colour_map):
    patch.set(facecolor=col, edgecolor="black", linewidth=3)
for element in ["whiskers", "caps", "medians"]:
    for artist in bp[element]:
        artist.set(color="black", linewidth=3)

# group labels centered
group_centres, start = [], 0
for g_len in group_lengths:
    group_centres.append(sum(positions[start:start+g_len]) / g_len)
    start += g_len

ax.set_xticks(group_centres)
ax.set_xticklabels(["Rb_K4me3_H23", "Ms_K4me3_H12"], fontsize=12)

ax.set_ylabel("FRiP per Barcode", fontsize=14)
ax.set_ylim(0, 1)

title = "FRiP per cell (barcode)"
if MIN_TOTAL_READS_PER_BARCODE and MIN_TOTAL_READS_PER_BARCODE > 0:
    title += f" | barcodes ≥ {MIN_TOTAL_READS_PER_BARCODE} reads"
ax.set_title(title, fontsize=12)

for spine in ax.spines.values():
    spine.set_linewidth(3)
ax.tick_params(width=3, length=6)

plt.tight_layout()
plt.savefig(OUT_PNG, dpi=300)
plt.show()

# Optional QC print
for c in rb_samples + ms_samples:
    vals = frip_per_condition.get(c, [])
    if len(vals) == 0:
        print(f"{c:28s}  n_cells=0")
    else:
        print(f"{c:28s}  n_cells={len(vals):6d}  median={np.median(vals):.3f}  mean={np.mean(vals):.3f}")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal, mannwhitneyu

# ────────────────────────────────────────────────────────────────
# 1) Define groups (edit if needed)
# ────────────────────────────────────────────────────────────────
rb_frip_group = [
    "Rb_K4me3_H23_CnT",
    "Rb_K4me3_H23_CoCnT_4X",
    "Rb_K4me3_H23_CoCnT_12X",
    "Rb_K4me3_H23_CoCnT_24X",
]
ms_frip_group = [
    "Ms_K4me3_H12_CnT",
    "Ms_K4me3_H12_CoCnT_4X",
    "Ms_K4me3_H12_CoCnT_12X",
    "Ms_K4me3_H12_CoCnT_24X",
]

FRIP_GROUPS = {
    "Rb_K4me3_H23": rb_frip_group,
    "Ms_K4me3_H12": ms_frip_group,
}

# ────────────────────────────────────────────────────────────────
# 2) BH/FDR helper
# ────────────────────────────────────────────────────────────────
def bh_fdr(pvals):
    pvals = np.asarray(pvals, dtype=float)
    n = pvals.size
    order = np.argsort(pvals)
    ranked = pvals[order]
    adj = ranked * n / (np.arange(1, n + 1))
    adj = np.minimum.accumulate(adj[::-1])[::-1]
    adj = np.clip(adj, 0, 1)
    out = np.empty_like(adj)
    out[order] = adj
    return out

# ────────────────────────────────────────────────────────────────
# 3) Per-group stats
# ────────────────────────────────────────────────────────────────
def within_group_frip_stats(frip_per_condition, group_name, conditions, alternative="two-sided"):
    present = [c for c in conditions if c in frip_per_condition and len(frip_per_condition[c]) > 0]
    if len(present) < 2:
        return None, None

    data = [np.asarray(frip_per_condition[c], dtype=float) for c in present]
    kw_stat, kw_p = kruskal(*data)

    rows = []
    pvals = []
    for i in range(len(present)):
        for j in range(i + 1, len(present)):
            a = data[i]
            b = data[j]

            u_stat, p = mannwhitneyu(a, b, alternative=alternative)

            # effect size: delta = 2U/(n1*n2) - 1  in [-1,1]
            n1, n2 = len(a), len(b)
            delta = (2 * u_stat) / (n1 * n2) - 1

            rows.append({
                "group": group_name,
                "cond_a": present[i],
                "cond_b": present[j],
                "n_a": n1,
                "n_b": n2,
                "median_a": float(np.median(a)),
                "median_b": float(np.median(b)),
                "U": float(u_stat),
                "p_raw": float(p),
                "effect_delta": float(delta),
            })
            pvals.append(p)

    p_adj = bh_fdr(pvals)
    for r, pa in zip(rows, p_adj):
        r["p_adj_BH"] = float(pa)

    pairwise_df = pd.DataFrame(rows).sort_values("p_adj_BH")
    omnibus_df = pd.DataFrame([{
        "group": group_name,
        "k": len(present),
        "conditions": ", ".join(present),
        "kruskal_stat": float(kw_stat),
        "kruskal_p": float(kw_p),
    }])

    return omnibus_df, pairwise_df

# ────────────────────────────────────────────────────────────────
# 4) Run all groups
# ────────────────────────────────────────────────────────────────
all_omni = []
all_pair = []

for gname, conds in FRIP_GROUPS.items():
    omni, pair = within_group_frip_stats(frip_per_condition, gname, conds)
    if omni is not None:
        all_omni.append(omni)
        all_pair.append(pair)

omnibus_df = pd.concat(all_omni, ignore_index=True) if all_omni else pd.DataFrame()
pairwise_df = pd.concat(all_pair, ignore_index=True) if all_pair else pd.DataFrame()

print("\nOmnibus (Kruskal–Wallis) per group:")
print(omnibus_df.to_string(index=False))

print("\nPairwise (Mann–Whitney U) within group (BH-adjusted):")
print(pairwise_df.to_string(index=False))

# Optional saves
# omnibus_df.to_csv("frip_omnibus_kruskal.csv", index=False)
# pairwise_df.to_csv("frip_pairwise_mwu_BH.csv", index=False)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal, mannwhitneyu

# ---- define your groups (using your updated names) ----
rb_samples_1 = [
    "Rb_K27me3_H1_CnT",
    "Rb_K27me3_H1_CoCnT_4X",
    "Rb_K27me3_H1_CoCnT_12X",
    "Rb_K27me3_H1_CoCnT_24X",
]

ms_samples_1 = [
    "Ms_K27me3_H11_CnT",
    "Ms_K27me3_H11_CoCnT_4X",
    "Ms_K27me3_H11_CoCnT_12X",
    "Ms_K27me3_H11_CoCnT_24X",
]

ms_samples_2 = [
    "Ms_K27me3_H18_CnT",
    "Ms_K27me3_H18_CoCnT_4X",
    "Ms_K27me3_H18_CoCnT_12X",
    "Ms_K27me3_H18_CoCnT_24X",
]

ms_samples_3 = [
    "Ms_K27me3_H19_CnT",
    "Ms_K27me3_H19_CoCnT_4X",
    "Ms_K27me3_H19_CoCnT_12X",
    "Ms_K27me3_H19_CoCnT_24X",
]

GROUPS = {
    "Rb_K27me3_H1": rb_samples_1,
    "Ms_K27me3_H11": ms_samples_1,
    "Ms_K27me3_H18": ms_samples_2,
    "Ms_K27me3_H19": ms_samples_3,
}

def bh_fdr(pvals):
    """Benjamini–Hochberg FDR correction. Returns adjusted p-values in original order."""
    pvals = np.asarray(pvals, dtype=float)
    n = pvals.size
    order = np.argsort(pvals)
    ranked = pvals[order]
    adj = ranked * n / (np.arange(1, n + 1))
    # enforce monotonicity
    adj = np.minimum.accumulate(adj[::-1])[::-1]
    adj = np.clip(adj, 0, 1)
    out = np.empty_like(adj)
    out[order] = adj
    return out

def within_group_stats(barcode_counts, group_name, conditions, alternative="two-sided"):
    """
    barcode_counts: dict condition -> pandas Series (barcode -> count), from your pipeline
    conditions: list of condition names to compare within this group
    """
    # keep only conditions that exist and have data
    present = [c for c in conditions if c in barcode_counts and len(barcode_counts[c]) > 0]
    if len(present) < 2:
        return None, None

    data = [barcode_counts[c].values for c in present]

    # omnibus test
    kw_stat, kw_p = kruskal(*data)

    # pairwise tests
    rows = []
    pvals = []
    for i in range(len(present)):
        for j in range(i + 1, len(present)):
            a = barcode_counts[present[i]].values
            b = barcode_counts[present[j]].values

            # Mann–Whitney U (nonparametric)
            u_stat, p = mannwhitneyu(a, b, alternative=alternative)

            # effect size (Cliff's delta-like using U)
            # delta = 2U/(n1*n2) - 1  (ranges [-1,1])
            n1, n2 = len(a), len(b)
            delta = (2 * u_stat) / (n1 * n2) - 1

            rows.append({
                "group": group_name,
                "cond_a": present[i],
                "cond_b": present[j],
                "n_a": n1,
                "n_b": n2,
                "median_a": float(np.median(a)),
                "median_b": float(np.median(b)),
                "U": float(u_stat),
                "p_raw": float(p),
                "effect_delta": float(delta),
            })
            pvals.append(p)

    # multiple testing correction within this group
    p_adj = bh_fdr(pvals)
    for r, pa in zip(rows, p_adj):
        r["p_adj_BH"] = float(pa)

    pairwise_df = pd.DataFrame(rows).sort_values("p_adj_BH")
    omnibus = pd.DataFrame([{
        "group": group_name,
        "k": len(present),
        "conditions": ", ".join(present),
        "kruskal_stat": float(kw_stat),
        "kruskal_p": float(kw_p),
    }])

    return omnibus, pairwise_df

# ---- run all groups ----
all_omnibus = []
all_pairwise = []

for gname, conds in GROUPS.items():
    omni, pair = within_group_stats(barcode_counts, gname, conds, alternative="two-sided")
    if omni is not None:
        all_omnibus.append(omni)
        all_pairwise.append(pair)

omnibus_df = pd.concat(all_omnibus, ignore_index=True) if all_omnibus else pd.DataFrame()
pairwise_df = pd.concat(all_pairwise, ignore_index=True) if all_pairwise else pd.DataFrame()

print("\nOmnibus (Kruskal–Wallis) per group:")
print(omnibus_df.to_string(index=False))

print("\nPairwise (Mann–Whitney U) within each group (BH-adjusted):")
print(pairwise_df.head(30).to_string(index=False))

# Optional saves
# omnibus_df.to_csv("reads_per_cell_omnibus_kruskal.csv", index=False)
# pairwise_df.to_csv("reads_per_cell_pairwise_mwu_BH.csv", index=False)
```

```python
# ────────────────────────────────────────────────────────────────
# FRiP per barcode for "conditions" where each condition = 1+ BED files
# ────────────────────────────────────────────────────────────────

import os
import numpy as np
import pandas as pd
import pybedtools
import matplotlib.pyplot as plt

# ────────────────────────────────────────────────────────────────
# 1) Inputs
# ────────────────────────────────────────────────────────────────

CONDITION_BEDS = {
    "Ms_K27me3_H19_CnT": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K27me3(H19)_CnT_041725.bed.gz",
    ],

    "Ms_K27me3_H19_CoCnT_4X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K27me3(H19)_wRb_K4me3(H2)_4XS_4XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K27me3_H1_4XS_4XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H23_4XS_4XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H24_4XS_4XM_060225.bed.gz",
    ],

    "Ms_K27me3_H19_CoCnT_12X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K27me3(H19)_wRb_K4me3(H2)_12XS_12XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K27me3_H1_12XS_12XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H23_12XS_12XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H24_12XS_12XM_060225.bed.gz",
    ],

    "Ms_K27me3_H19_CoCnT_24X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Ms_K27me3(H19)_wRb_K4me3(H2)_24XS_24XM_041725.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K27me3_H1_24XS_24XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H23_24XS_24XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Ms_K27me3_H19_wRb_K4me3_H24_24XS_24XM_060225.bed.gz",
    ],

    "Rb_K27me3_H1_CnT": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Rb_K27me3(H1)_CnT_041725.bed.gz",
    ],

    "Rb_K27me3_H1_CoCnT_4X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Rb_K27me3(H1)_wMs_K4me3(H12)_4XS_4XM_041725.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K27me3_H19_4XS_4XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K4me3_H12_4XS_4XM_060225.bed.gz",
    ],

    "Rb_K27me3_H1_CoCnT_12X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Rb_K27me3(H1)_wMs_K4me3(H12)_12XS_12XM_041725.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K27me3_H19_12XS_12XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K4me3_H12_12XS_12XM_060225.bed.gz",
    ],

    "Rb_K27me3_H1_CoCnT_24X": [
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001695_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/AE_Hs_Rb_K27me3(H1)_wMs_K4me3(H12)_24XS_24XM_041725.bed.gz",
        #"/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K27me3_H19_24XS_24XM_060225.bed.gz",
        "/varidata/research/projects/janssens/Janssens_Lab_Data/PR001743_sciCoCnT_JAND_DNA_ALIGNMENT_WELLID_BARCODES/BED/HA_Hs_Rb_K27me3_H1_wMs_K4me3_H12_24XS_24XM_060225.bed.gz",
    ],
}

# Peak set (single bed)
PEAK_BED = "/varidata/research/projects/janssens/CoCnT_Data_Analysis/Combined_K4me3_K27me3_beds/K27me3_H1_H19_CnT_CoCnT_Combined.01.stringent.bed"

# Column indices (0-based)
CHR_COL, START_COL, END_COL, BARCODE_COL = 0, 1, 2, 3

# Optional filter for stability: only keep barcodes with >= N total reads across all files in the condition
MIN_TOTAL_READS_PER_BARCODE = 250   # set 0 to disable

# Plot grouping/order
rb_samples = [
    "Rb_K27me3_H1_CnT",
    "Rb_K27me3_H1_CoCnT_4X",
    "Rb_K27me3_H1_CoCnT_12X",
    "Rb_K27me3_H1_CoCnT_24X",
]


ms_samples = [
    "Ms_K27me3_H19_CnT",
    "Ms_K27me3_H19_CoCnT_4X",
    "Ms_K27me3_H19_CoCnT_12X",
    "Ms_K27me3_H19_CoCnT_24X",
]

colour_map = (
    ["#4D4D4D", "#ADD8E6", "#1E90FF", "#104E8B"] +  # Rb
    ["#4D4D4D", "#FFB3FF", "#FF00FF", "#8B008B"]    # Ms
)

OUT_PNG = (
    "/varidata/research/projects/janssens/CoCnT_Data_Analysis/Figure_Panels/"
    "K27me3_H1_H19_FRIP_per_cell_boxplot_ALLREADS_combinedConditions.png"
)

# ────────────────────────────────────────────────────────────────
# 2) Prepare peak BedTool once
# ────────────────────────────────────────────────────────────────
peak_bt = pybedtools.BedTool(PEAK_BED).sort().merge()

# ────────────────────────────────────────────────────────────────
# 3) Helpers
# ────────────────────────────────────────────────────────────────

def load_bed_mincols(path, usecols=(0,1,2,3)):
    return pd.read_csv(path, sep="\t", header=None, compression="infer", usecols=list(usecols))

def count_total_reads_per_barcode(df, barcode_col=3):
    return df[barcode_col].value_counts()

def count_in_peaks_per_barcode(df, peak_bt, chr_col=0, start_col=1, end_col=2, barcode_col=3):
    # Create BedTool from 4-column df (chr, start, end, barcode)
    bt = pybedtools.BedTool.from_dataframe(df[[chr_col, start_col, end_col, barcode_col]])
    in_peak = bt.intersect(peak_bt, wa=True)
    if len(in_peak) == 0:
        return pd.Series(dtype=int)
    in_df = in_peak.to_dataframe(names=[chr_col, start_col, end_col, barcode_col])
    return in_df[barcode_col].value_counts()

def frip_for_condition(bed_paths):
    """
    Returns list of FRiP values (one per barcode), after summing reads across all bed_paths.
    FRiP = (reads in peaks across all files) / (total reads across all files) for each barcode.
    """
    total_counts = None
    peak_counts = None

    for p in bed_paths:
        df = load_bed_mincols(p, usecols=(CHR_COL, START_COL, END_COL, BARCODE_COL))
        if df.empty:
            continue

        tc = count_total_reads_per_barcode(df, barcode_col=BARCODE_COL)
        pc = count_in_peaks_per_barcode(df, peak_bt,
                                        chr_col=CHR_COL, start_col=START_COL, end_col=END_COL, barcode_col=BARCODE_COL)

        total_counts = tc if total_counts is None else total_counts.add(tc, fill_value=0)
        peak_counts  = pc if peak_counts  is None else peak_counts.add(pc, fill_value=0)

    if total_counts is None or len(total_counts) == 0:
        return []

    total_counts = total_counts.astype(int)
    peak_counts = (peak_counts if peak_counts is not None else pd.Series(dtype=float)).astype(float)

    # Optional barcode filter on total reads (across all combined files)
    if MIN_TOTAL_READS_PER_BARCODE and MIN_TOTAL_READS_PER_BARCODE > 0:
        keep = total_counts[total_counts >= MIN_TOTAL_READS_PER_BARCODE].index
        total_counts = total_counts.loc[keep]
        peak_counts = peak_counts.reindex(keep, fill_value=0)

    # FRiP per barcode (aligned)
    frip = peak_counts.reindex(total_counts.index, fill_value=0).divide(total_counts)

    # Clean numeric + bound
    frip = frip.replace([np.inf, -np.inf], np.nan).dropna()
    frip = frip.clip(0, 1)

    return frip.tolist()

# ────────────────────────────────────────────────────────────────
# 4) Compute FRiP per barcode for each condition (combining files)
# ────────────────────────────────────────────────────────────────
frip_per_condition = {}
for cond, paths in CONDITION_BEDS.items():
    if isinstance(paths, str):
        paths = [paths]
    frip_per_condition[cond] = frip_for_condition(paths)

# ────────────────────────────────────────────────────────────────
# 5) Build data & box-plot (gapped layout)
# ────────────────────────────────────────────────────────────────

data_rb = [frip_per_condition[c] for c in rb_samples]
data_ms = [frip_per_condition[c] for c in ms_samples]

all_data = data_rb + data_ms
group_lengths = [len(data_rb), len(data_ms)]

# positions with 1 blank gap
positions, current, gap = [], 0, 1
for g_len in group_lengths:
    positions.extend(range(current, current + g_len))
    current += g_len + gap

fig, ax = plt.subplots(figsize=(6, 6))
bp = ax.boxplot(
    all_data,
    positions=positions,
    widths=0.8,
    patch_artist=True,
    showfliers=False
)

# color + style
for patch, col in zip(bp["boxes"], colour_map):
    patch.set(facecolor=col, edgecolor="black", linewidth=3)
for element in ["whiskers", "caps", "medians"]:
    for artist in bp[element]:
        artist.set(color="black", linewidth=3)

# group labels centered
group_centres, start = [], 0
for g_len in group_lengths:
    group_centres.append(sum(positions[start:start+g_len]) / g_len)
    start += g_len

ax.set_xticks(group_centres)
ax.set_xticklabels(["Rb_K4me3_H23", "Ms_K4me3_H12"], fontsize=12)

ax.set_ylabel("FRiP per Barcode", fontsize=14)
ax.set_ylim(0, 1)

title = "FRiP per cell (barcode)"
if MIN_TOTAL_READS_PER_BARCODE and MIN_TOTAL_READS_PER_BARCODE > 0:
    title += f" | barcodes ≥ {MIN_TOTAL_READS_PER_BARCODE} reads"
ax.set_title(title, fontsize=12)

for spine in ax.spines.values():
    spine.set_linewidth(3)
ax.tick_params(width=3, length=6)

plt.tight_layout()
plt.savefig(OUT_PNG, dpi=300)
plt.show()

# Optional QC print
for c in rb_samples + ms_samples:
    vals = frip_per_condition.get(c, [])
    if len(vals) == 0:
        print(f"{c:28s}  n_cells=0")
    else:
        print(f"{c:28s}  n_cells={len(vals):6d}  median={np.median(vals):.3f}  mean={np.mean(vals):.3f}")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal, mannwhitneyu

# ────────────────────────────────────────────────────────────────
# 1) Define groups (edit if needed)
# ────────────────────────────────────────────────────────────────
rb_frip_group = [
    "Rb_K27me3_H1_CnT",
    "Rb_K27me3_H1_CoCnT_4X",
    "Rb_K27me3_H1_CoCnT_12X",
    "Rb_K27me3_H1_CoCnT_24X",
]
ms_frip_group = [
    "Ms_K27me3_H19_CnT",
    "Ms_K27me3_H19_CoCnT_4X",
    "Ms_K27me3_H19_CoCnT_12X",
    "Ms_K27me3_H19_CoCnT_24X",
]

FRIP_GROUPS = {
    "Rb_K27me3_H1": rb_frip_group,
    "Ms_K27me3_H19": ms_frip_group,
}

# ────────────────────────────────────────────────────────────────
# 2) BH/FDR helper
# ────────────────────────────────────────────────────────────────
def bh_fdr(pvals):
    pvals = np.asarray(pvals, dtype=float)
    n = pvals.size
    order = np.argsort(pvals)
    ranked = pvals[order]
    adj = ranked * n / (np.arange(1, n + 1))
    adj = np.minimum.accumulate(adj[::-1])[::-1]
    adj = np.clip(adj, 0, 1)
    out = np.empty_like(adj)
    out[order] = adj
    return out

# ────────────────────────────────────────────────────────────────
# 3) Per-group stats
# ────────────────────────────────────────────────────────────────
def within_group_frip_stats(frip_per_condition, group_name, conditions, alternative="two-sided"):
    present = [c for c in conditions if c in frip_per_condition and len(frip_per_condition[c]) > 0]
    if len(present) < 2:
        return None, None

    data = [np.asarray(frip_per_condition[c], dtype=float) for c in present]
    kw_stat, kw_p = kruskal(*data)

    rows = []
    pvals = []
    for i in range(len(present)):
        for j in range(i + 1, len(present)):
            a = data[i]
            b = data[j]

            u_stat, p = mannwhitneyu(a, b, alternative=alternative)

            # effect size: delta = 2U/(n1*n2) - 1  in [-1,1]
            n1, n2 = len(a), len(b)
            delta = (2 * u_stat) / (n1 * n2) - 1

            rows.append({
                "group": group_name,
                "cond_a": present[i],
                "cond_b": present[j],
                "n_a": n1,
                "n_b": n2,
                "median_a": float(np.median(a)),
                "median_b": float(np.median(b)),
                "U": float(u_stat),
                "p_raw": float(p),
                "effect_delta": float(delta),
            })
            pvals.append(p)

    p_adj = bh_fdr(pvals)
    for r, pa in zip(rows, p_adj):
        r["p_adj_BH"] = float(pa)

    pairwise_df = pd.DataFrame(rows).sort_values("p_adj_BH")
    omnibus_df = pd.DataFrame([{
        "group": group_name,
        "k": len(present),
        "conditions": ", ".join(present),
        "kruskal_stat": float(kw_stat),
        "kruskal_p": float(kw_p),
    }])

    return omnibus_df, pairwise_df

# ────────────────────────────────────────────────────────────────
# 4) Run all groups
# ────────────────────────────────────────────────────────────────
all_omni = []
all_pair = []

for gname, conds in FRIP_GROUPS.items():
    omni, pair = within_group_frip_stats(frip_per_condition, gname, conds)
    if omni is not None:
        all_omni.append(omni)
        all_pair.append(pair)

omnibus_df = pd.concat(all_omni, ignore_index=True) if all_omni else pd.DataFrame()
pairwise_df = pd.concat(all_pair, ignore_index=True) if all_pair else pd.DataFrame()

print("\nOmnibus (Kruskal–Wallis) per group:")
print(omnibus_df.to_string(index=False))

print("\nPairwise (Mann–Whitney U) within group (BH-adjusted):")
print(pairwise_df.to_string(index=False))

# Optional saves
# omnibus_df.to_csv("frip_omnibus_kruskal.csv", index=False)
# pairwise_df.to_csv("frip_pairwise_mwu_BH.csv", index=False)
```
